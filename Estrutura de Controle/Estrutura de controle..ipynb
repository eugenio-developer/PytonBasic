<font color="#0D9DF0"><h3><B>ESTRUTURA DE CONTROLE</b></h3></font>
***

As estruturas de controle são blocos de código que permitem controlar o fluxo de execução de um programa, ou seja, decidir o que será executado, quando e quantas vezes.

#### Para que servem?
Com elas, o programa deixa de ser "linear" (linha por linha) e passa a tomar decisões, repetir ações ou reagir a erros de forma inteligente.

#### Como funcionam no Python?
Em Python, usamos indentação (recuo) para definir blocos de código dentro dessas estruturas. Diferente de outras linguagens, não usamos chaves {}, mas sim dois pontos : e recuo com espaços ou tabulação.


<font color="yellow">Principais Tipos</font>
<center>

| Tipo de Estrutura          | Exemplo          | Para quê serve?                                |
|----------------------------|------------------|------------------------------------------------|
| `if / elif / else`         | Decisão          | Executa um bloco condicionalmente              |
| `for` e `while`            | Repetição        | Executa algo várias vezes                      |
| `break`, `continue`, `pass`| Controle de Fluxo| Interrompe, pula ou ignora partes do código    |
| `try / except`             | Tratamento de Erros | Evita que o programa pare em caso de erro  |

</center>


#### Estruturas Condicionais (Decisões)
***
As estruturas condicionais permitem que o programa tome decisões com base em uma condição. Usamos:

`if`: se a condição for verdadeira

`elif`: senão se...

`else`: senão (caso contrário)

#### Estruturas de Repetição (Laços de Loop)
***
Permitem executar blocos de código repetidamente. Os dois principais em Python são:

`for`: para um número definido de repetições

`while`: enquanto uma condição for verdadeira

#### Comandos de Controle de Fluxo
***
Controlam o comportamento dentro dos loops.

`break`: encerra o loop imediatamente

`continue`: pula para a próxima iteração

`pass`: faz nada (placeholder)

***
