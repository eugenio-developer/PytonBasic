<font color="##188038"><h4><B> Listas (`list`)</b></h4></font>
***

Listas são coleções ordenadas e mutáveis. Podemos adicionar, remover ou alterar elementos.

Usos no mundo real:

- Listas de tarefas: Aplicativos de produtividade armazenam as tarefas do usuário.

- Carrinho de compras: E-commerces armazenam os itens adicionados pelo cliente.

- Histórico de navegação: Navegadores mantêm as páginas visitadas em listas.

- Séries de dados: Em ciência de dados, listas armazenam valores contínuos ou registros.

````python 
# Exemplos
tarefas = ["Estudar Python", "Fazer exercícios", "Assistir aula"]
carrinho = ["camiseta", "calça", "tênis"]
````


##### Funções/métodos úteis:
<center>

| Função/Método       | Descrição                            |
|---------------------|----------------------------------------|
| `append()`          | Adiciona um item ao final da lista     |
| `insert()`          | Insere um item em uma posição específica |
| `remove()`          | Remove a primeira ocorrência de um item |
| `pop()`             | Remove e retorna um item pelo índice   |
| `sort()` / `reverse()` | Ordena / inverte os elementos       |
| `len()`             | Retorna o tamanho da lista             |

</center>


##### Exemplo prático:


`append()` — Adiciona um item ao final da lista

`insert()` — Insere um item em uma posição específica

`remove()` — Remove a primeira ocorrência de um item

`pop()` — Remove e retorna um item pelo índice

`sort()` — Ordena os elementos

`reverse()` — Inverte a ordem dos elementos

`len()` — Retorna o número de elementos
