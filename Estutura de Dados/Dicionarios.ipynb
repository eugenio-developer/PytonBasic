<font color="##188038"><h4><B> Dicionários (`dict`)</b></h4></font>
***
Dicionários armazenam dados em pares **chave:valor**. São mutáveis e não ordenados até o Python 3.7+ (agora mantêm ordem de inserção).

Usos no mundo real:

- Cadastro de usuários: Nome, idade, email associados a uma chave única.

- Resposta de APIs: Dados estruturados vêm em formato de dicionário (JSON).

- Configurações de sistema: Parâmetros nomeados e seus respectivos valores.

- Dados de banco de dados em memória.

```python 
# Exemplo
usuario = {
    "nome": "Maria",
    "email": "maria@gmail.com",
    "idade": 30
}
```

##### Funções/métodos úteis:
<center>


| Função/Método       | Descrição                             |
|---------------------|-----------------------------------------|
| `keys()`            | Retorna as chaves                      |
| `values()`          | Retorna os valores                     |
| `items()`           | Retorna pares (chave, valor)           |
| `get()`             | Retorna o valor de uma chave           |
| `update()`          | Atualiza o dicionário com novos pares  |
| `pop()`             | Remove um item pela chave              |

##### Exemplo prático:

`keys()` — Retorna todas as chaves

`values()` — Retorna todos os valores

`items()` — Retorna todos os pares (chave, valor)

`get()` — Retorna o valor de uma chave

`update()` — Atualiza ou adiciona pares chave:valor

`pop()` — Remove uma chave e retorna seu valor

#
<font color="##188038"><h4><B>Trabalhando com Dicionários que Armazenam Múltiplos Valores em Python</b></h4></font>
***

Em Python, os **dicionários** são estruturas poderosas para armazenar dados relacionados por meio de pares `chave: valor`. Porém, em muitas situações do mundo real, cada chave pode estar associada a **mais de um valor** — como uma pessoa com vários telefones, um aluno com várias notas, ou um usuário com diversos cursos.

Para isso, usamos outras estruturas (como **listas**, **tuplas** ou até **outros dicionários**) como valor de cada chave, criando dicionários **aninhados ou compostos**.

Essa abordagem permite representar dados de forma mais rica, organizada e acessível.


<font color="yellow"><h7>Exemplo 1: Lista como valor</h7></font>

Um cliente pode ter vários pedidos:


***


<font color="yellow"><h7>Exemplo 2: Dicionário dentro de dicionário</h7></font>

Informações mais completas de um usuário, como nome, idade e cursos feitos:

***

<font color="yellow"><h7>Exemplo 3: Tuplas como valor</h7></font>

Localizações em coordenadas (imutáveis):

***


<font color="yellow"><h7>Exemplo 4: Lista de dicionários</h7></font>

Cada item da lista representa um registro mais detalhado:

***
