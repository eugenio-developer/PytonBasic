<font color="##188038"><h4><B>Tuplas (`tuple`)</b></h4></font>
***

Tuplas são coleções ordenadas e **imutáveis**. Após criadas, seus valores não podem ser alterados.


Usos no mundo real:

- Coordenadas geográficas: Latitude e longitude não mudam.

- Informações fixas de cadastro: Nome, CPF e data de nascimento.

- Resultados de funções que retornam múltiplos valores.

- Jogos: Posições fixas de objetos no mapa (x, y).

```python 
# Exemplos
coordenadas = (23.5505, -46.6333)
usuario = ("João", "123.456.789-00", "10/10/1990")
```

##### Funções/métodos úteis:
<center>

| Função/Método       | Descrição                             |
|---------------------|-----------------------------------------|
| `count()`           | Conta quantas vezes um valor aparece   |
| `index()`           | Retorna o índice da primeira ocorrência |
| `len()`             | Retorna o tamanho da tupla             |

</center>

##### Exemplo prático:

`count()` — Conta quantas vezes um valor aparece

`index()` — Retorna a posição da primeira ocorrência

`len()` — Retorna o número de elementos
