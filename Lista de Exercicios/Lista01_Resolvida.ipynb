```python
a = 10.0
if isinstance(a,int) :
    print('inteiro')
elif isinstance(a,float):
    print ('float')
```

5. Peça ao usuário que digite um número e mostre se ele é par ou ímpar.

```python
try:
    numero = int(input('digite um Numero Inteiro'))
except:
    print('entrada invalida!')
else :
    resultado = 'par' if numero % 2 == 0 else 'impar'
    print(resultado)
```
