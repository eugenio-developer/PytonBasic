## Parte 1: Tipos Primitivos em Python

**Objetivo:** Praticar o uso e manipulação dos tipos primitivos: `int`, `float`, `bool`, `str`.

### Exercícios


1. Declare uma variável do tipo `int` com qualquer valor e imprima seu tipo usando `type()`.

```python
idade = 29
type(idade)
```

2. Crie duas variáveis, uma `int` e outra `float`. Some-as e exiba o resultado e seu tipo.

```python
cenoura = 15
abobora = 12.8
soma = cenoura + abobora
print(soma)
type(soma)
```

3. Converta uma string `"123"` para o tipo `int` e imprima o resultado.

```python
codigo = '123'
codigoConvertido = int(codigo)
print(codigoConvertido)
type(codigoConvertido)
```

4. Verifique se a variável `a = 10.0` é do tipo `int` ou `float` usando `isinstance()`.

```python
a = 10.0
if isinstance(a,int) :
    print('inteiro')
elif isinstance(a,float):
    print ('float')
```

5. Peça ao usuário que digite um número e mostre se ele é par ou ímpar.

```python
try:
    numero = int(input('digite um Numero Inteiro'))
except:
    print('entrada invalida!')
else :
    resultado = 'par' if numero % 2 == 0 else 'impar'
    print(resultado)
```

6. Crie uma variável booleana com valor `True`, depois inverta seu valor e imprima.

```python
controle = True

print(not controle)
```

7. Transforme o número `3.14159` em uma string e mostre o tipo da nova variável.

```python
numero = 3.14159
resultado = str(numero)
type(resultado)
```

8. Escreva um programa que receba dois números do usuário e exiba a média entre eles.

```python
primeiro = int(input('digite um numero'))
segundo = int(input('digite um segundo numero'))
media = (primeiro + segundo) / 2
print(media)
```

9. Multiplique uma string por um número e mostre o resultado. Ex: `"oi" * 3`

```python
print('oi' * 3)
```

10. Receba uma idade digitada pelo usuário e verifique se ela é maior ou igual a 18.

```python
idade = int(input('Digite a sua idade: '))
resultado = 'maior de 18 anos' if idade >= 18 else 'menor de 18 anos'
print(resultado)
```

11. Receba uma nota (float) de um aluno e mostre se ele foi aprovado (nota >= 6).

```python
nota = float(input('digite a nota do aluno: '))
resultado = 'APROVADO!' if nota > 6 else 'REPROVADO'
print(resultado)
```

12. Use `input()` para receber o nome do usuário e exiba uma saudação.

```python
nome = input('digite seu nome: ')
print('ola', nome, ', bem vindo!')
```

13. Crie três variáveis com tipos diferentes (`int`, `float`, `str`) e imprima seus valores e tipos.

```python
x = 3
y = 3.14
z = '3.14'

print(x,y,z)

print(type(x))
print(type(y))
print(type(z))
```

## Parte 2: Funções em Python

**Objetivo:** Desenvolver a capacidade de criar, chamar e usar funções com parâmetros e retorno.

### Exercícios

1. Crie uma função que imprime "Olá, mundo!".


```python
def hello():
    print('hello world')

hello()
```

2. Escreva uma função que recebe um nome como parâmetro e imprime uma saudação.

```python
def saudacao(nome):
    print('olá', nome, 'bem vindo!')

saudacao('raquel')
```

3. Crie uma função que receba dois números como parâmetros e retorne a soma deles.


```python
def soma(x,y):
    return sum(x,y)
```

4. Crie uma função que retorne o quadrado de um número.


```python
def quadrado(numero):
    return numero ** 2
```

5. Faça uma função que verifique se um número é par.


```python
def par(numero):
    if numero % 2 == 0:
        return True
    else :
        return False
```

6. Crie uma função que receba três notas e retorne a média.


```python
def media (nota1, nota2, nota3):
    return (nota1 + nota2 + nota3) / 3
```

7. Faça uma função que retorne o maior entre dois números.


```python
def maior(a,b):
    if a > b:
        return a
    else:
        return b
```

8. Crie uma função que receba um texto e o retorne em letras maiúsculas.


```python
def maiuscula(texto):
    return texto.upper()

print(maiuscula('ola mundo'))
```

9. Faça uma função que receba uma lista de números e retorne a soma deles.


```python
def soma(numeros):
    return sum(numeros)
```

10. Escreva uma função que retorne a quantidade de caracteres de uma string.


```python
def tamanho(texto):
    return len(texto)
```

11. Escreva uma função que receba uma lista de nomes e imprima cada um com uma saudação.

```python
def saudar_nomes(lista_nomes):
    list(map(lambda nome: print("Olá,", nome, '!'), lista_nomes))
```
