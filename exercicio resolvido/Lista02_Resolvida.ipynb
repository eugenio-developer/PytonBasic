## Parte 3: Estruturas de Controle

**Objetivo:** Praticar o uso de estruturas condicionais (`if`, `elif`, `else`) e de repetição (`for`, `while`).

### Exercícios
1. Peça ao usuário um número e diga se ele é positivo, negativo ou zero.

```python
try:
    numero = int(input('digite um numero: '))
    if numero > 0:
        print('positivo!')
    elif numero < 0:
        print('negativo!')
    else:
        print('o numero é zero!')
except:
    print('entrada invalida!')
```

2. Crie um programa que receba três números e diga qual é o maior deles.

```python
a, b, c = map(int, input("digite 3 numeros: ").split())
print(max(a,b,c))
```

3. Escreva um código que peça ao usuário sua idade e exiba se ele é criança (até 12), adolescente (13 a 17), adulto (18 a 59) ou idoso (60+).

```python
idade = int(input('digite a sua idade: '))

if idade <= 12:
    print('criança!')
elif idade <= 17:
    print('adolescente!')
elif idade <= 59:
    print('adulto!')
else:
    print('idoso!')
```

4. Faça um programa que imprima os números de 1 a 10 usando `for`.

```python
for i in range(1,11):
    print(i)
```

5. Faça o mesmo que o anterior, mas usando `while`.

```python
contador = 1
while contador < 11:
    print(contador)
    contador+=1
    
```

6. Escreva um programa que leia 5 números e mostre a soma deles.

```python
soma = 0
for i in range(5):
    numero = int(input('digite um numero'))
    soma += numero
print(soma)
```

7. Crie um laço que peça números ao usuário até que ele digite 0. No final, exiba a soma total.

```python
soma = 0
while True:
    numero = int(input('digite um numero: '))
    if numero == 0:
        print(soma)
        break
    soma += numero
```

8. Escreva um código que conte quantos números pares existem entre 1 e 100.

```python
contador = 0
for i in range(1,101):
    if i % 2 == 0:
        contador += 1
print(contador)
```
