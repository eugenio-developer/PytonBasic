```python
try:
    numero = int(input('digite um numero: '))
    if numero > 0:
        print('positivo!')
    elif numero < 0:
        print('negativo!')
    else:
        print('o numero é zero!')
except:
    print('entrada invalida!')
```

```python
idade = int(input('digite a sua idade: '))

if idade <= 12:
    print('criança!')
elif idade <= 17:
    print('adolescente!')
elif idade <= 59:
    print('adulto!')
else:
    print('idoso!')
```

```python
soma = 0
while True:
    numero = int(input('digite um numero: '))
    if numero == 0:
        print(soma)
        break
    soma += numero
```
