<font color="#0D9DF0"><h3><B>O QUE SÃO TIPOS DE DADOS EM PYTHON?</b></h3></font>
***


Todo valor em Python possui um tipo. Esses tipos determinam o que você pode fazer com esses valores, como somar, fatiar, iterar, entre outras operações.

Python é uma linguagem de tipagem dinâmica, ou seja, você não precisa declarar o tipo da variável — o próprio Python reconhece automaticamente.


##### Por que é importante saber os tipos? 
Saber os tipos de dados ajuda a:

- Usar corretamente as variáveis

- Fazer operações válidas (ex: não dá pra somar texto com número direto)
 
- Controlar melhor a entrada/saída de dados
 
- Trabalhar com listas, dicionários, e outras estruturas com segurança
 
- Fazer cast/conversões de tipos (int(), str(), etc.)


##### Quais são os tipos mais usados?

<center>

| Tipo de Dado   | Exemplo              | Para que serve?                                      |
|----------------|----------------------|------------------------------------------------------|
| `int`          | `10`, `-3`, `0`      | Números inteiros, usados em contagem, operações     |
| `float`        | `3.14`, `-2.5`, `0.0`| Números decimais, usados para medidas, porcentagens |
| `str`          | `"Olá"` ou `'Python'`| Texto, nomes, mensagens, etc                         |
| `bool`         | `True`, `False`      | Lógica condicional, verificações                    |


</center>
