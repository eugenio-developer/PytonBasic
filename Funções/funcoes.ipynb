<font color="#0D9DF0"><h3><B>O QUE É UMA FUNÇÃO EM PYTHON?</b></h3></font>
***

Uma função é um bloco de código que executa uma tarefa específica e pode ser reutilizado sempre que necessário.
Ela é como uma "mini-máquina" que recebe uma entrada (se necessário), processa algo, e pode devolver uma saída (retorno).

Vantagens de usar funções:
- Evita repetição de código

- Organiza melhor o programa

- Facilita a manutenção

- Permite reutilização em vários lugares

#### Contexto de Uso de Funções

***

Funções são usadas em quase todo programa Python, desde scripts simples até grandes sistemas. Alguns exemplos:

- Sites e sistemas web (funções que calculam preços, geram relatórios, etc)
- Análises de dados (funções que processam e limpam dados)
- Jogos (funções para movimentar personagens, calcular pontuação)
- Inteligência Artificial (funções para treinar modelos, fazer previsões)


#### O que é Escopo?
***

Escopo se refere ao "alcance" ou "visibilidade" de uma variável, ou seja, onde no código ela pode ser acessada.

Existem dois principais tipos de escopo:

1. Escopo Local
- A variável só existe dentro de uma função. Ela não pode ser usada fora dela.

2. Escopo Global
- A variável é definida fora de qualquer função, então pode ser acessada em todo o código (inclusive dentro de funções, se necessário).

##### Dica!
Evite modificar variáveis globais dentro de funções. Isso pode deixar o código confuso e difícil de manter.

#### Retorno de Funções
***
O retorno é o resultado final de uma função. Usamos a palavra-chave `return` para devolver valores ao final da função.

A função pode retornar:

- Um único valor

- Múltiplos valores (em forma de tupla)
 
- Nenhum valor (ela apenas executa ações, como print())

#### Boas práticas ao usar parâmetros e retorno
***
- Dê nomes claros aos parâmetros, de acordo com o que representam.
- Use valores padrão sempre que possível para tornar a função mais flexível.
- Sempre que a função produzir um resultado, use return (e não apenas print). 
- Retorne apenas o necessário, evitando retornar dados em excesso.

#### Boas Práticas ao Criar Funções
***
Seguir boas práticas ajuda seu código a ser mais limpo, organizado, reutilizável e fácil de entender. 

Veja algumas dicas importantes:

1. Use nomes significativos

Dê nomes claros e diretos às funções e variáveis:

```python 
# Bom exemplo de variáveis
def calcular_desconto(valor, percentual):

# Exemplo ruim (não é claro)
def cd(v, p): 
```

 2. Mantenha as funções curtas

Funções devem fazer uma única tarefa bem definida. Se estiver fazendo muitas coisas, talvez seja hora de dividir em funções menores.

 3. Evite repetir código

Se você usa o mesmo bloco de código várias vezes, transforme-o em uma função. Isso facilita manutenções futuras.

 4. Prefira funções puras

Funções que recebem dados e retornam resultados sem depender de variáveis externas são mais fáceis de testar e entender.

 5. Documente suas funções

Mesmo que pareça óbvio para você hoje, no futuro (ou para outra pessoa), pode não ser. Um bom docstring evita confusão.

 6. Evite efeitos colaterais desnecessários

Não altere variáveis globais dentro da função, a menos que realmente precise. Prefira retornar valores e tratar fora.



#### Exemplo final com boas práticas

#### Exemplo de uso
